## 5.1 Principles of Good Tables

### ğŸ¯ Why Use Tables?

Tables are best for:
- âœ… Presenting exact numerical values
- âœ… Showing multiple variables simultaneously
- âœ… Allowing precise comparisons

#### 2. **Formatting**

**DO:**
- âœ… Use horizontal lines only (top, below headers, bottom)
- âœ… Left-align text, right-align numbers
- âœ… Include units in column headers
- âœ… Use consistent decimal places
- âœ… Use meaningful abbreviations (explained in notes)
- âœ… Number tables consecutively (Table 1, Table 2...)

**DON'T:**
- âŒ Use vertical lines
- âŒ Use colors or shading (unless essential)
- âŒ Repeat units in every cell
- âŒ Use vague titles ("Results", "Data")
- âŒ Include too many decimal places

---

#### 3. **Content**

**What to Include:**
- Sample sizes (n)
- Measures of central tendency (mean, median)
- Measures of variability (SD, SE, range)
- Statistical test results (if appropriate)
- Significance indicators (*, **, ***)

**Statistical Notation:**
- Mean Â± SD (standard deviation)
- Mean Â± SE (standard error)
- Mean (range: min-max)
- Median (IQR: Q1-Q3)

**Significance Indicators:**
- * p < 0.05
- ** p < 0.01
- *** p < 0.001
- NS = not significant

## 6.1 Why Cite?

### ğŸ¯ Purposes of Citation

1. **Give Credit** - Acknowledge others' work
2. **Avoid Plagiarism** - Show what's yours vs. others'
3. **Support Claims** - Provide evidence
4. **Allow Verification** - Readers can check sources
5. **Show Scholarship** - Demonstrate you've done homework

---

## 6.2 When to Cite

### âœ… ALWAYS Cite When:

- Using someone else's ideas, theories, or findings
- Quoting directly (exact words)
- Paraphrasing (putting in your own words)
- Using data, statistics, or figures from others
- Referring to specific studies or methods
- Using established protocols (e.g., "DNA was extracted using the CTAB method (Doyle & Doyle, 1987)")

### âŒ DON'T Cite:

- Common knowledge ("Fish have gills", "DNA is double-stranded")
- Your own original data
- Your own ideas and interpretations
- General dictionaries or encyclopedias (use specialized sources instead)

---

## 6.3 Major Citation Styles

### ğŸ“– Three Main Styles in Biology

#### 1. **APA Style** (American Psychological Association)
- Used in: Psychology, Education, some Biology journals
- In-text: (Author, Year)
- Focus on: Publication date

#### 2. **Harvard Style** (Author-Date)
- Used in: UK universities, many Biology journals
- In-text: (Author Year)
- Very similar to APA

#### 3. **Vancouver Style** (Numbered)
- Used in: Medical journals, some Science journals
- In-text: Superscript numbers [1]
- References numbered in order of appearance

**For BSc/MSc Zoology:** Usually **APA** or **Harvard**

**Check with:** Your supervisor, university guidelines, or target journal

---

## 6.4 APA Style (7th Edition) - Most Common

### ğŸ“ In-Text Citations

#### **One Author:**
- Narrative: Kumar (2020) found that earthworms accumulate lead...
- Parenthetical: Earthworms accumulate lead (Kumar, 2020).

#### **Two Authors:**
- Narrative: Patel and Singh (2021) reported...
- Parenthetical: The species was first described (Patel & Singh, 2021).

#### **Three or More Authors:**
- First citation: Sharma et al. (2019)...
- All subsequent: Sharma et al. (2019)...
- Note: "et al." means "and others"

#### **Multiple Sources:**
- (Kumar, 2020; Patel & Singh, 2021; Sharma et al., 2019)
- Order: Alphabetically by first author

#### **Same Author, Different Years:**
- (Kumar, 2019, 2020, 2021)

#### **Same Author, Same Year:**
- (Kumar, 2020a, 2020b)
- Order: Alphabetically by title

#### **Direct Quote:**
- Short (<40 words): Kumar (2020) stated that "earthworms are excellent biomonitors" (p. 45).
- Long (â‰¥40 words): Block quote, indented, no quotation marks

---

### ğŸ“š Reference List Format

**General Rules:**
- Alphabetical order by first author's surname
- Hanging indent (first line flush left, subsequent lines indented)
- Double-spaced
- Title: "References" (centered, not bold)

## 6.7 Avoiding Plagiarism

### âš ï¸ What is Plagiarism?

**Plagiarism** = Using others' words or ideas without proper acknowledgment

**Types:**

1. **Direct Plagiarism** - Copying word-for-word without quotes/citation
2. **Mosaic Plagiarism** - Mixing copied phrases with your own
3. **Paraphrasing Plagiarism** - Rewording without citation
4. **Self-Plagiarism** - Reusing your own previous work without disclosure

---

### âœ… How to Avoid Plagiarism

#### **1. Quote Correctly**

**Original:**
"Earthworms play a crucial role in soil health by decomposing organic matter and improving soil structure."

**Correct Quote:**
Kumar (2020) stated that "earthworms play a crucial role in soil health by decomposing organic matter and improving soil structure" (p. 45).

---

#### **2. Paraphrase Properly**

**Original:**
"Earthworms play a crucial role in soil health by decomposing organic matter and improving soil structure."

---

#### **3. When in Doubt, Cite!**

**Better to over-cite than under-cite**

## 7.1 Principles of Effective Data Visualization

### ğŸ¨ The Golden Rules

1. **Clarity Over Complexity**
   - Simple is better than fancy
   - Remove unnecessary elements
   - Each figure should tell ONE story

2. **Accuracy**
   - Never distort data
   - Start y-axis at zero (for bar charts)
   - Use appropriate scales

3. **Honesty**
   - Show uncertainty (error bars)
   - Don't hide unfavorable data
   - Acknowledge limitations

4. **Accessibility**
   - Colorblind-friendly palettes
   - Large enough text
   - Clear labels

---

## 7.2 Choosing the Right Plot Type

### ğŸ“ˆ Decision Tree

| Your Data | Use This Plot |
|-----------|---------------|
| **One continuous variable** | Histogram, Box plot |
| **Compare groups (continuous)** | Box plot, Violin plot |
| **Compare means** | Bar chart (with error bars) |
| **Relationship between 2 variables** | Scatter plot |
| **Time series** | Line graph |
| **Proportions/percentages** | Pie chart, Stacked bar |
| **Distributions comparison** | Density plot, Violin plot |
| **Correlations (many variables)** | Heatmap |

---

## 7.3 Publication-Quality Figure Elements

### âœ… Every Figure Must Have:

1. **Clear Axes Labels**
   - Variable name
   - Units in parentheses
   - Example: "Body Weight (g)"

2. **Informative Title/Caption**
   - Goes BELOW figure (unlike tables)
   - Self-explanatory
   - Explains symbols, abbreviations

3. **Legend** (if multiple groups)
   - Clear, unambiguous
   - Inside or outside plot area

4. **Error Bars** (for means)
   - Specify what they represent (SD, SE, 95% CI)
   - Example: "Error bars represent Â± SD"

5. **Statistical Annotations** (if relevant)
   - Show significance: *, **, ***
   - Brackets for pairwise comparisons

---

## 7.4 Color Theory for Science

### ğŸ¨ Color Guidelines

**DO:**
- âœ… Use colorblind-friendly palettes
- âœ… Ensure sufficient contrast
- âœ… Use colors consistently (same color = same thing)
- âœ… Consider cultural meanings

**DON'T:**
- âŒ Use red-green combinations (colorblindness)
- âŒ Use rainbow colors (poor for quantitative data)
- âŒ Use too many colors (max 6-7)

**Recommended Palettes:**
- **Qualitative:** Colorblind-safe (matplotlib 'tab10', seaborn 'colorblind')
- **Sequential:** Viridis, Blues, Greens
- **Diverging:** RdBu (Red-Blue), for data with meaningful center

## 7.5 Figure Captions

### âœï¸ Writing Good Captions

**Structure:**
```

         [Symbol explanations].
```

**Example:**
```
Figure 1. Comparison of fish growth rates under different feeding regimes. 
Mean (Â± SD) weekly weight gain of Labeo rohita fed with three different
```

````python
# Creating Publication-Quality Figures

def create_publication_figures():
    """
    Demonstrate creating different types of publication-quality figures
    """
    
    np.random.seed(42)
    
    # Generate sample data
    control = np.random.normal(450, 50, 30)
    treatment_a = np.random.normal(520, 50, 30)
    treatment_b = np.random.normal(580, 55, 30)
    
    # Create figure with multiple subplots
    fig = plt.figure(figsize=(16, 12))
    
    # 1. Box Plot with significance
    ax1 = plt.subplot(2, 3, 1)
    data_box = [control, treatment_a, treatment_b]
    bp = ax1.boxplot(data_box, labels=['Control', 'Treatment A', 'Treatment B'],
                     patch_artist=True, widths=0.6)
    
    # Color the boxes
    colors = ['#3498db', '#2ecc71', '#e74c3c']
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)
        patch.set_alpha(0.7)
    
    ax1.set_ylabel('Fish Weight (g)', fontsize=11, fontweight='bold')
    ax1.set_title('A. Comparison of Fish Weights', fontsize=12, fontweight='bold', loc='left')
    ax1.grid(axis='y', alpha=0.3)
    
    # Add significance brackets
    y_max = max([d.max() for d in data_box]) + 20
    ax1.plot([1, 3], [y_max, y_max], 'k-', linewidth=1.5)
    ax1.text(2, y_max + 10, '***', ha='center', fontsize=14)
    
    # 2. Bar Chart with Error Bars
    ax2 = plt.subplot(2, 3, 2)
    means = [np.mean(d) for d in data_box]
    sds = [np.std(d) for d in data_box]
    x_pos = np.arange(len(means))
    
    bars = ax2.bar(x_pos, means, yerr=sds, capsize=5, 
                   color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)
    ax2.set_xticks(x_pos)
    ax2.set_xticklabels(['Control', 'Treatment A', 'Treatment B'])
    ax2.set_ylabel('Mean Weight (g)', fontsize=11, fontweight='bold')
    ax2.set_title('B. Mean Weights with SD', fontsize=12, fontweight='bold', loc='left')
    ax2.grid(axis='y', alpha=0.3)
    
    # Add value labels on bars
    for i, (bar, mean) in enumerate(zip(bars, means)):
        ax2.text(bar.get_x() + bar.get_width()/2, mean + sds[i] + 10,
                f'{mean:.0f}', ha='center', fontsize=9, fontweight='bold')
    
    # 3. Scatter Plot with Regression
    ax3 = plt.subplot(2, 3, 3)
    length = np.random.normal(25, 3, 90)
    weight = 0.5 * length**2 + np.random.normal(0, 20, 90)
    
    ax3.scatter(length, weight, alpha=0.6, s=50, c='#3498db', edgecolors='black', linewidth=0.5)
    
    # Add regression line
    z = np.polyfit(length, weight, 2)
    p = np.poly1d(z)
    x_line = np.linspace(length.min(), length.max(), 100)
    ax3.plot(x_line, p(x_line), 'r--', linewidth=2, label='Fitted curve')
    
    # Calculate RÂ²
    r_squared = 1 - (np.sum((weight - p(length))**2) / np.sum((weight - weight.mean())**2))
    ax3.text(0.05, 0.95, f'RÂ² = {r_squared:.3f}', transform=ax3.transAxes,
            fontsize=10, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    ax3.set_xlabel('Fish Length (cm)', fontsize=11, fontweight='bold')
    ax3.set_ylabel('Fish Weight (g)', fontsize=11, fontweight='bold')
    ax3.set_title('C. Length-Weight Relationship', fontsize=12, fontweight='bold', loc='left')
    ax3.legend(fontsize=9)
    ax3.grid(alpha=0.3)
    
    # 4. Violin Plot
    ax4 = plt.subplot(2, 3, 4)
    parts = ax4.violinplot(data_box, positions=[1, 2, 3], showmeans=True, showmedians=True)
    
    for i, pc in enumerate(parts['bodies']):
        pc.set_facecolor(colors[i])
        pc.set_alpha(0.7)
    
    ax4.set_xticks([1, 2, 3])
    ax4.set_xticklabels(['Control', 'Treatment A', 'Treatment B'])
    ax4.set_ylabel('Fish Weight (g)', fontsize=11, fontweight='bold')
    ax4.set_title('D. Distribution Comparison', fontsize=12, fontweight='bold', loc='left')
    ax4.grid(axis='y', alpha=0.3)
    
    # 5. Time Series
    ax5 = plt.subplot(2, 3, 5)
    weeks = np.arange(0, 13)
    growth_ctrl = 50 + weeks * 5 + np.random.normal(0, 3, len(weeks))
    growth_trt = 50 + weeks * 8 + np.random.normal(0, 3, len(weeks))
    
    ax5.plot(weeks, growth_ctrl, 'o-', linewidth=2, markersize=6, 
            label='Control', color='#3498db')
    ax5.plot(weeks, growth_trt, 's-', linewidth=2, markersize=6,
            label='Treatment', color='#2ecc71')
    ax5.fill_between(weeks, growth_ctrl - 5, growth_ctrl + 5, alpha=0.2, color='#3498db')
    ax5.fill_between(weeks, growth_trt - 5, growth_trt + 5, alpha=0.2, color='#2ecc71')
    
    ax5.set_xlabel('Time (weeks)', fontsize=11, fontweight='bold')
    ax5.set_ylabel('Cumulative Weight Gain (g)', fontsize=11, fontweight='bold')
    ax5.set_title('E. Growth Over Time', fontsize=12, fontweight='bold', loc='left')
    ax5.legend(fontsize=9)
    ax5.grid(alpha=0.3)
    
    # 6. Heatmap
    ax6 = plt.subplot(2, 3, 6)
    
    # Correlation matrix
    variables = ['Length', 'Weight', 'Temperature', 'pH', 'DO']
    corr_matrix = np.random.rand(5, 5)
    corr_matrix = (corr_matrix + corr_matrix.T) / 2  # Make symmetric
    np.fill_diagonal(corr_matrix, 1)  # Diagonal = 1
    
    im = ax6.imshow(corr_matrix, cmap='RdBu_r', vmin=-1, vmax=1, aspect='auto')
    
    # Add colorbar
    cbar = plt.colorbar(im, ax=ax6)
    cbar.set_label('Correlation Coefficient', fontsize=10)
    
    # Set ticks
    ax6.set_xticks(np.arange(len(variables)))
    ax6.set_yticks(np.arange(len(variables)))
    ax6.set_xticklabels(variables, rotation=45, ha='right')
    ax6.set_yticklabels(variables)
    
    # Add values
    for i in range(len(variables)):
        for j in range(len(variables)):
            text = ax6.text(j, i, f'{corr_matrix[i, j]:.2f}',
                          ha="center", va="center", color="black", fontsize=8)
    
    ax6.set_title('F. Correlation Matrix', fontsize=12, fontweight='bold', loc='left')
    
    plt.tight_layout()
    plt.savefig('/mnt/user-data/outputs/publication_figures.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print("\n" + "="*80)
    print("ğŸ“Š PUBLICATION-QUALITY FIGURES CREATED")
    print("="*80)
    print("\nâœ… All figures include:")
    print("   â€¢ Clear axis labels with units")
    print("   â€¢ Informative titles (A, B, C... for multi-panel)")
    print("   â€¢ Appropriate error bars or confidence intervals")
    print("   â€¢ Colorblind-friendly colors")
    print("   â€¢ High resolution (300 DPI)")
    print("   â€¢ Grid for easier reading")
    print("\nğŸ’¾ Figure saved as: publication_figures.png (300 DPI)")
    print("\nğŸ“ Caption Template:")
    print("\nFigure 1. Comprehensive analysis of fish growth experiment. ")
    print("(A) Box plots comparing weight distributions across treatments (n=30 per group). ")
    print("(B) Mean weights with standard deviation error bars. ")
    print("(C) Scatter plot showing length-weight relationship with quadratic fit (RÂ²=0.XX). ")
    print("(D) Violin plots illustrating full distribution shapes. ")
    print("(E) Growth trajectories over 12 weeks, shaded areas represent Â±SD. ")
    print("(F) Correlation heatmap of measured variables. ")
    print("Asterisks indicate significance: ***p < 0.001.")
    print("="*80)

# Create the figures
create_publication_figures()
````
