## 4.2 Writing Each Section

### ğŸ“ TITLE

**Purpose:** Attract readers, describe study accurately

**Good Title Characteristics:**
- âœ… Specific and informative
- âœ… Includes key variables
- âœ… Concise (12-20 words ideal)
- âœ… No abbreviations
- âœ… No unnecessary words ("A Study of...", "An Investigation into...")

**Examples:**

| Poor Title | Good Title |
|------------|------------|
| "Fish Study" | "Effect of Water Temperature on Growth Rate of Labeo rohita in Aquaculture Ponds" |
| "A Study on Earthworms" | "Earthworm Species Composition as Biomonitor for Heavy Metal Contamination in Mining Areas of Western Odisha" |
| "Research on Butterflies" | "Impact of Forest Fragmentation on Butterfly Diversity in Western Odisha: A Comparative Analysis" |

---

### ğŸ“„ ABSTRACT

**Purpose:** Self-contained summary of entire study

**Structure (usually 150-300 words):**

1. **Background** (1-2 sentences)
   - Context and importance
   
2. **Objective** (1 sentence)
   - What you aimed to do
   
3. **Methods** (2-3 sentences)
   - How you did it (briefly)
   
4. **Results** (3-4 sentences)
   - Key findings with numbers
   
5. **Conclusion** (1-2 sentences)
   - Main conclusion and significance

**Example Abstract:**

```
Heavy metal contamination from mining activities poses significant 
environmental risks in Western Odisha, necessitating cost-effective 
biomonitoring approaches. This study investigated the potential of 
earthworm species composition as a biomonitor for heavy metal 
contamination in mining areas. Earthworm surveys were conducted at 
10 sites (5 mining, 5 agricultural) using standardized quadrat

DNA barcoding (COI gene), and soil heavy metal concentrations (Pb, 
Cd, Cu) were measured using Atomic Absorption Spectroscopy. Mining 
sites showed significantly higher metal concentrations (mean Pb:

distinctly different earthworm assemblages. Metal-tolerant species 
(Eisenia fetida, Lumbricus rubellus) dominated mining sites (78% of 
total abundance), while agricultural sites harbored diverse species 
assemblages (Shannon diversity: 2.3 Â± 0.2 vs 0.9 Â± 0.1, p < 0.001). 
Earthworm species composition reliably indicates heavy metal

### ğŸ“– 1. INTRODUCTION

**Purpose:** Set the context, justify the study

**Structure (Funnel approach - broad to specific):**

```
        BROAD CONTEXT
             â†“
        Narrow to topic
             â†“
      Literature review
             â†“
        Knowledge gap
             â†“
    Your study (specific)
```

**What to Include:**

1. **Opening** (1-2 paragraphs)
   - General importance of topic
   - Why should anyone care?

2. **Literature Review** (3-5 paragraphs)
   - What's already known
   - Key previous studies
   - Different perspectives

3. **Knowledge Gap** (1 paragraph)
   - What's NOT known
   - Why it matters

4. **Study Rationale** (1 paragraph)
   - How your study addresses the gap
   - Regional/local importance

5. **Objectives & Hypotheses** (1 paragraph)
   - Clear statement of aims
   - Testable predictions

**Writing Tips:**
- âœ… Use past tense for previous studies
- âœ… Use present tense for general facts
- âœ… Cite all claims properly
- âœ… Be critical, not just descriptive
- âŒ Don't include your results here
- âŒ Don't overcite (not every sentence needs citation)

### ğŸ”¬ 2. MATERIALS AND METHODS

**Purpose:** Allow others to replicate your study

**Golden Rule:** Another researcher should be able to repeat your study exactly from this section!

**Sub-sections:**

#### 2.1 Study Area
- Location (coordinates)
- Geographic description
- Climate
- Vegetation/habitat type
- Include map if relevant

#### 2.2 Study Design
- Type of study (experimental, observational, etc.)
- Duration
- Sampling design

#### 2.3 Sample Collection
- Exactly how samples were collected
- Equipment used
- Sample size
- When collected (dates, time of day)

#### 2.4 Laboratory Procedures
- Step-by-step protocols
- Equipment specifications
- Chemical concentrations
- Standards used

#### 2.5 Data Analysis
- Statistical tests used
- Software (with version number)
- Significance level (Î± = 0.05)
- How assumptions were tested

**Writing Tips:**
- âœ… Use past tense ("were collected", "was measured")
- âœ… Be specific with numbers and units
- âœ… Use subheadings for clarity
- âœ… Include equipment model numbers
- âœ… Cite established methods
- âŒ Don't include results
- âŒ Don't justify choices here (that's for Discussion)

**Example Sentence:**

âŒ **Vague:** "Earthworms were collected from soil."

âœ… **Specific:** "Earthworms were collected using 0.5m Ã— 0.5m quadrats placed randomly at each site (n=30 per site). Soil was excavated to 30cm depth and hand-sorted in the field. Specimens were preserved in 70% ethanol and transported to the laboratory within 24 hours."

### ğŸ“Š 3. RESULTS

**Purpose:** Present findings objectively, without interpretation

**Golden Rule:** JUST THE FACTS! No interpretation, no speculation.

**Structure:**

1. **Order by objectives** - Address each objective sequentially

2. **General to specific** - Start with overview, then details

3. **Use visuals** - Tables and figures to present data

4. **Text supports visuals** - Don't just repeat what's in tables/figures

**What to Include:**

- âœ… Descriptive statistics (mean Â± SD)
- âœ… Statistical test results (test statistic, p-value)
- âœ… Effect sizes when relevant
- âœ… Clear statement of significant/non-significant results
- âœ… Refer to tables and figures

**What NOT to Include:**

- âŒ Interpretation ("because...", "this suggests...")
- âŒ Speculation ("probably", "might be")
- âŒ Comparison with other studies
- âŒ Implications

**Example Result Statements:**

âŒ **Bad (has interpretation):**
"Fish in Pond B grew faster, probably because the fertilizer increased food availability."

âœ… **Good (just facts):**
"Fish in Pond B showed significantly higher growth rates (5.2 Â± 0.8 g/week) compared to Pond A (3.1 Â± 0.6 g/week; t-test, t = 8.34, df = 58, p < 0.001)."

**Writing Tips:**
- âœ… Use past tense
- âœ… Report statistics in standard format
- âœ… Be precise with numbers
- âœ… Use "significant" only for statistical significance
- âœ… Include both positive and negative results

---

### ğŸ’¬ 4. DISCUSSION

**Purpose:** Interpret results, explain meaning

**Structure (Inverted funnel - specific to broad):**

```
    Your findings (specific)
             â†“
      Interpret results
             â†“
    Compare with literature
             â†“
      Broader implications
             â†“
        GENERAL CONTEXT
```

**What to Include:**

1. **Opening** (1 paragraph)
   - Restate main findings briefly
   - Did results support hypotheses?

2. **Detailed Interpretation** (3-5 paragraphs)
   - Explain WHAT results mean
   - Explain WHY you got these results
   - Mechanisms, processes

3. **Comparison with Literature** (2-3 paragraphs)
   - Do your results agree or disagree with others?
   - Why might there be differences?
   - What's novel about your findings?

4. **Limitations** (1 paragraph)
   - Acknowledge weaknesses honestly
   - How might they affect conclusions?
   - Don't apologize excessively!

5. **Implications** (1-2 paragraphs)
   - Practical applications
   - Theoretical contributions
   - Policy relevance

6. **Future Research** (1 paragraph)
   - What questions remain?
   - Specific suggestions

**Writing Tips:**
- âœ… Use present tense for general conclusions
- âœ… Use past tense when referring to your results
- âœ… Be critical but balanced
- âœ… Cite relevant literature
- âŒ Don't introduce new results
- âŒ Don't overgeneralize
- âŒ Don't ignore conflicting evidence

---

### ğŸ¯ 5. CONCLUSIONS

**Purpose:** Summarize key findings and their significance

**Keep it:**
- Short (1-2 paragraphs or bullet points)
- Specific (answer your objectives)
- Impactful (broader significance)

**Example:**

```

indicates heavy metal contamination in mining areas of Western Odisha. 
The dominance of metal-tolerant species (E. fetida, L. rubellus) in 
contaminated sites provides a cost-effective biomonitoring tool for 
environmental assessment. These findings support the implementation of 
earthworm-based biomonitoring programs for mining regions, contributing 
to more sustainable environmental management practices.
