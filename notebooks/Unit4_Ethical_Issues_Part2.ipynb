## 5.1 Understanding Royalties

### üíµ What is a Royalty?

**Royalty** = Payment made to IP owner for use of their property

**Common in research:**
- Patents licensed to companies
- Technology transfer
- Software licensing
- Book publishing

---

### üìä How Royalties Work

**Typical Arrangement:**

```
Patent/IP ‚Üí Licensed to Company ‚Üí Product Sales ‚Üí Royalty Payment
                                                         ‚Üì
                                  Split between: Institution + Inventor(s)
```

**Example Royalty Structure:**

| Recipient | Typical Share |
|-----------|---------------|
| **University** | 40-60% |
| **Department** | 0-20% |
| **Inventor(s)** | 30-50% |

**Rates vary by:**
- Institution policy
- Type of invention
- Negotiated terms

---

### üí° Real Example: University Patent

**Scenario:**
- Professor develops new fish vaccine
- University files patent
- Licensed to pharmaceutical company
- Royalty: 5% of net sales
- Product generates ‚Çπ10 crore in sales

**Royalty calculation:**
```
Total royalty = ‚Çπ10 crore √ó 5% = ‚Çπ50 lakhs

Distribution (example):
University:   ‚Çπ25 lakhs (50%)
Department:   ‚Çπ5 lakhs (10%)
Inventor:     ‚Çπ20 lakhs (40%)
```

## 5.2 Technology Transfer

### üîÑ What is Technology Transfer?

**Definition:** Moving research discoveries from lab ‚Üí commercial use

**Process:**

```
1. INVENTION
   Researcher makes discovery
   ‚Üì
2. DISCLOSURE
   Report to Technology Transfer Office (TTO)
   ‚Üì
3. EVALUATION
   TTO assesses commercial potential
   ‚Üì
4. PROTECTION
   File patent/copyright/trademark
   ‚Üì
5. MARKETING
   Find industry partner
   ‚Üì
6. LICENSING
   Negotiate agreement
   ‚Üì
7. COMMERCIALIZATION
   Product development & sales
   ‚Üì
8. ROYALTIES
   Revenue sharing
```

---

### üè¢ Types of Licenses

#### 1. **Exclusive License**
- Only one licensee can use the technology
- Higher royalty rates
- More attractive to companies (protects investment)

#### 2. **Non-Exclusive License**
- Multiple licensees possible
- Lower royalty rates
- Broader dissemination

#### 3. **Field-Specific License**
- Exclusive for specific application
- Example: diagnostic use vs. therapeutic use

## 5.3 Open Access vs Traditional Publishing

### üìñ Two Publishing Models

#### **Traditional (Subscription-Based)**

**How it works:**
- Author submits paper ‚Üí Publisher peer reviews ‚Üí Published
- Author transfers copyright to publisher
- Readers/libraries pay subscription to access
- Author pays nothing (usually)

**Pros:**
- ‚úÖ No cost to author
- ‚úÖ Established prestige
- ‚úÖ Quality control

**Cons:**
- ‚ùå Limited access (behind paywall)
- ‚ùå Author loses copyright
- ‚ùå Can't share freely
- ‚ùå Slower dissemination

---

#### **Open Access (OA)**

**How it works:**
- Author submits paper ‚Üí Publisher peer reviews ‚Üí Published
- Author retains copyright (usually)
- Readers access FREE online
- Author pays Article Processing Charge (APC)

**Pros:**
- ‚úÖ Free access for everyone
- ‚úÖ Higher visibility
- ‚úÖ Greater citations
- ‚úÖ Faster dissemination
- ‚úÖ Author retains rights
- ‚úÖ Meets funder requirements

**Cons:**
- ‚ùå Author pays (‚Çπ50,000 - ‚Çπ3,00,000)
- ‚ùå Risk of predatory journals
- ‚ùå Variable quality

---

### üÜì Types of Open Access

#### **Gold OA**
- Published in full OA journal
- Immediate free access
- Author pays APC

#### **Green OA**
- Published in traditional journal
- Author self-archives in repository
- Usually after embargo period (6-24 months)
- No APC

#### **Hybrid OA**
- Traditional journal with OA option
- Author pays to make their article OA
- Controversial ("double dipping")

---

### ‚ö†Ô∏è Predatory Journals

**Warning Signs:**
- Spam email invitations
- Promise rapid publication
- Poor English on website
- No clear peer review process
- Unrealistic scope ("all sciences")
- Suspicious contact info
- Low/no APC (then hidden fees)

**How to Check:**
- ‚úÖ Check if journal indexed in PubMed/Web of Science/Scopus
- ‚úÖ Look for Impact Factor from legitimate source
- ‚úÖ Check Directory of Open Access Journals (DOAJ)
- ‚úÖ Ask your supervisor
- ‚úÖ Check journal's editorial board (real people?)

**Resources:**
- Think.Check.Submit (www.thinkchecksubmit.org)
- Beall's List (archived)
- UGC-CARE List (India)

## 6.1 Types of Plagiarism

### üìö Complete Classification

#### **1. Direct Plagiarism (Clone)**

**What it is:**
- Word-for-word copying
- No quotation marks
- No citation

**Severity:** ‚ö†Ô∏è‚ö†Ô∏è‚ö†Ô∏è VERY SERIOUS

**Example:**

**Original (Kumar, 2020):**
"Earthworms are ecosystem engineers that modify soil structure through their burrowing activities and contribute to nutrient cycling by decomposing organic matter."

**Plagiarized:**
"Earthworms are ecosystem engineers that modify soil structure through their burrowing activities and contribute to nutrient cycling by decomposing organic matter."

---

#### **2. Mosaic Plagiarism (Patchwork)**

**What it is:**
- Mix of copied phrases and your words
- Substituting synonyms
- Rearranging but keeping structure

**Severity:** ‚ö†Ô∏è‚ö†Ô∏è‚ö†Ô∏è VERY SERIOUS

**Example:**

**Original (Kumar, 2020):**
"Earthworms are ecosystem engineers that modify soil structure through their burrowing activities and contribute to nutrient cycling by decomposing organic matter."

**Plagiarized (Mosaic):**
"Earthworms are considered ecosystem engineers because they alter soil structure via their digging behavior and help with nutrient cycling through breaking down organic material."

**Problem:** Just synonym replacement, same structure, no citation!

---

#### **3. Paraphrasing Plagiarism**

**What it is:**
- Rewriting in own words
- BUT no citation
- Still using someone's ideas

**Severity:** ‚ö†Ô∏è‚ö†Ô∏è SERIOUS

**Example:**

**Original (Kumar, 2020):**
"Earthworms are ecosystem engineers that modify soil structure through their burrowing activities and contribute to nutrient cycling by decomposing organic matter."

**Plagiarized (Good paraphrase, no citation):**
"Through their tunneling behaviors and decomposition activities, earthworms play an important role in maintaining soil health and facilitating the movement of nutrients through ecosystems."

**Problem:** Good paraphrase, but NO CITATION! Still plagiarism!

**Correct Version:**
"Through their tunneling behaviors and decomposition activities, earthworms play an important role in maintaining soil health and facilitating the movement of nutrients through ecosystems (Kumar, 2020)."

---

#### **4. Self-Plagiarism**

**What it is:**
- Reusing your own previous work
- Without disclosure/citation
- Presenting old work as new

**Severity:** ‚ö†Ô∏è‚ö†Ô∏è SERIOUS

**Examples:**
- Submitting same paper to two journals
- Using methods section from old paper without citing
- Republishing same data as "new" study
- Copying from your thesis into a paper without disclosure

**Why it's wrong:**
- Inflates publication count
- Wastes reviewers' time
- Violates copyright (if transferred)
- Misleads readers about novelty

**When is it OK?**
- ‚úÖ Short standard methods (with citation to your earlier work)
- ‚úÖ Converting thesis chapters to papers (with disclosure)
- ‚úÖ Review articles citing your own work
- ‚úÖ When explicitly allowed by publisher

---

#### **5. Accidental Plagiarism**

**What it is:**
- Unintentional failure to cite
- Poor paraphrasing
- Forgetting quotation marks
- Lost track of sources

**Severity:** ‚ö†Ô∏è SERIOUS (ignorance is not an excuse)

**Common causes:**
- Poor note-taking
- Not recording sources while researching
- Confusion about what needs citation
- Insufficient understanding of rules

**Prevention:**
- Use citation management software
- Record sources immediately
- Mark direct quotes clearly in notes
- When in doubt, cite!

---

#### **6. Source-Based Plagiarism**

**What it is:**
- Citing sources you didn't actually read
- Making up citations
- Incorrect citations

**Severity:** ‚ö†Ô∏è‚ö†Ô∏è SERIOUS

**Examples:**
- ‚ùå Copying citation from someone else without checking original
- ‚ùå Citing paper you couldn't access
- ‚ùå Inventing citations to support claim
- ‚ùå Wrong author names or dates

## 6.2 Plagiarism Detection

### üîç How Plagiarism is Detected

#### **1. Software Tools**

**Commercial:**
- **Turnitin** (most widely used)
  - Compares against database of billions of documents
  - Shows similarity percentage
  - Highlights matching text

- **iThenticate** (for journals)
  - Professional version of Turnitin
  - Used by publishers

- **Grammarly Premium**
  - Includes plagiarism checker
  - Limited database

**Free:**
- **Quetext** (limited free checks)
- **Duplichecker** (basic detection)
- **Google Scholar** (manual search)

---

#### **2. Manual Detection**

**Red Flags for Reviewers/Examiners:**
- Sudden change in writing style
- Inconsistent formatting
- Mix of British and American English
- Outdated references in "new" work
- Too polished for student level
- Suspicious phrases (easy to Google)

---

### üìä Understanding Similarity Reports

**Turnitin Similarity Index:**

| Score | Interpretation | Action |
|-------|----------------|--------|
| **0-15%** | ‚úÖ Usually acceptable | OK if citations proper |
| **15-25%** | ‚ö†Ô∏è Review needed | Check if quotes/citations |
| **25-50%** | ‚ö†Ô∏è‚ö†Ô∏è Concerning | Likely problematic |
| **>50%** | ‚ö†Ô∏è‚ö†Ô∏è‚ö†Ô∏è Serious issue | Major plagiarism likely |

**Important:** 
- Percentage alone doesn't tell full story
- Must examine WHAT is similar
- Common phrases are OK
- Properly cited quotes are OK
- Methods sections naturally similar

## 6.3 How to Avoid Plagiarism

### ‚úÖ Best Practices

#### **1. During Research Phase**

**Good Note-Taking:**
```
Source: Kumar, R. (2020). Heavy metal accumulation...
DOI: 10.1234/jeb.2020.234

DIRECT QUOTE (p. 235):
"Earthworms are ecosystem engineers that modify soil structure"

MY PARAPHRASE:
Kumar found that earthworms change soil properties through 
burrowing (Kumar, 2020, p. 235).

MY IDEAS/THOUGHTS:
This could explain the results I saw in my experiment...
```

**Key:** Clearly mark what's quoted, what's paraphrased, what's yours!

---

#### **2. During Writing Phase**

**Strategy:**
1. Read source material thoroughly
2. Close the source
3. Write in your own words from memory
4. Add citation
5. Go back and verify accuracy

**DON'T:**
- Have source open while writing
- Copy-paste then "modify"
- Write with source in front of you

---

#### **3. When to Cite**

**ALWAYS cite:**
- ‚úÖ Direct quotes
- ‚úÖ Paraphrased ideas
- ‚úÖ Specific facts/data
- ‚úÖ Others' theories/hypotheses
- ‚úÖ Others' research findings
- ‚úÖ Borrowed methods
- ‚úÖ Images/figures/tables from others

**DON'T cite:**
- ‚ùå Common knowledge ("fish have gills")
- ‚ùå Your own ideas
- ‚ùå Your own data
- ‚ùå Your own analysis
- ‚ùå Well-known historical facts

---

#### **4. Use Technology**

**Citation Management:**
- Zotero, Mendeley, EndNote
- Automatically formats citations
- Prevents missing references
- Organizes sources

**Check Before Submission:**
- Run through plagiarism checker
- Review similarity report
- Fix any issues

## 7.1 Research Misconduct

### üö´ The "Big Three" - FFP

#### **1. Fabrication**

**Definition:** Making up data or results

**Examples:**
- ‚ùå Creating fake experimental data
- ‚ùå Inventing observations that never happened
- ‚ùå Generating images/graphs from non-existent experiments
- ‚ùå Making up survey responses

**Real Case:**
- Researcher claimed to observe 500 earthworms
- Actually only collected 50
- Made up data for the other 450
- Detected when results too perfect (no natural variation)

---

#### **2. Falsification**

**Definition:** Manipulating data or processes to get desired results

**Examples:**
- ‚ùå Changing data points to fit hypothesis
- ‚ùå Selectively removing "outliers" without justification
- ‚ùå Manipulating images (enhance bands, remove unwanted signals)
- ‚ùå Reporting only favorable results, hiding unfavorable
- ‚ùå Changing experimental conditions without reporting

**Example:**
```
Raw data:        Modified data:
12, 15, 18,     12, 15, 18,
14, 32, 16,  ‚Üí  14, [DELETED], 16,  ‚Üê Removed "outlier" that didn't fit
13, 17, 15      13, 17, 15
```

**When is outlier removal OK?**
- ‚úÖ When you have clear reason (measurement error, equipment malfunction)
- ‚úÖ When documented and reported
- ‚úÖ When pre-specified in protocol

**When is it falsification?**
- ‚ùå Just because point doesn't fit your hypothesis
- ‚ùå Without documentation
- ‚ùå After seeing the results

## 7.2 Other Forms of Misconduct

### üìã Questionable Research Practices (QRPs)

**Not as severe as FFP, but still unethical:**

#### **1. HARKing**
**H**ypothesizing **A**fter **R**esults are **K**nown

**What it is:**
- Collect data without hypothesis
- Find interesting pattern
- Pretend you predicted it beforehand

**Why it's wrong:**
- Misleads about confirmatory vs. exploratory research
- Inflates apparent success rate
- Post-hoc hypotheses need testing!

**How to do it right:**
- ‚úÖ Be honest: "We unexpectedly found..."
- ‚úÖ Label exploratory analyses as exploratory
- ‚úÖ Suggest hypotheses for future testing

---

#### **2. P-Hacking**

**What it is:**
- Trying many analyses until you get p < 0.05
- Only reporting the "significant" one

**Examples:**
- Testing multiple variables, only report significant ones
- Trying different statistical tests until one gives p < 0.05
- Adding participants until you get significance
- Removing "outliers" to get significance

**Why it's wrong:**
- Inflates Type I error (false positives)
- Multiple comparisons not accounted for
- Results won't replicate

**How to avoid:**
- ‚úÖ Pre-register analysis plan
- ‚úÖ Report all analyses performed
- ‚úÖ Use appropriate corrections for multiple comparisons
- ‚úÖ Report negative results too

---

#### **3. Cherry-Picking**

**What it is:**
- Selecting only favorable data/results
- Ignoring contradictory evidence

**Examples:**
- ‚ùå Reporting 3 successful experiments, hiding 7 failed ones
- ‚ùå Only citing papers that agree with your results
- ‚ùå Using only best-looking images in figure

**Why it's wrong:**
- Misrepresents true state of evidence
- Biases scientific literature
- Wastes others' resources (false leads)

---

#### **4. Salami Slicing**

**What it is:**
- Dividing one study into many small papers
- "Least publishable unit" approach

**Example:**
- Study measured 10 variables in fish
- Published as 10 separate papers, each on one variable
- Rather than one comprehensive paper

**Why it's problematic:**
- Inflates publication count
- Fragments literature
- Wastes reviewers' time
- Hides full picture from readers

**When is it OK?**
- ‚úÖ Different aims/hypotheses
- ‚úÖ Different audiences
- ‚úÖ Papers truly independent
- ‚úÖ Cross-referenced and disclosed

## 7.3 Authorship Ethics

### ‚úçÔ∏è Who Should Be an Author?

**ICMJE Criteria (International Committee of Medical Journal Editors):**

**ALL FOUR required:**

1. ‚úÖ **Substantial contributions** to:
   - Conception or design, OR
   - Data acquisition, analysis, or interpretation

2. ‚úÖ **Drafting or revising** the manuscript
   - Critical intellectual input

3. ‚úÖ **Final approval** of version to be published

4. ‚úÖ **Accountability** for all aspects
   - Can explain and defend work

---

### ‚ùå Who Should NOT Be Author?

**"Honorary" or "Gift" Authorship:**
- ‚ùå Department head (who didn't contribute)
- ‚ùå Funding provider
- ‚ùå Equipment provider
- ‚ùå Person who gave general advice

**"Ghost" Authorship:**
- ‚ùå Company employee who wrote paper but isn't listed
- ‚ùå Professional writer hired by company

**Coercive Authorship:**
- ‚ùå Supervisor demands inclusion despite no contribution
- ‚ùå Senior researcher trades authorship

---

### üìã Author Order

**Typical Convention:**

```
First Author     - Did most of the work (often student)
                   Corresponding author (handles communication)
                   
Middle Authors   - Contributed substantially
                   Order by decreasing contribution
                   
Last Author      - Supervised the work (often PI/supervisor)
                   Senior author
                   Provides resources/guidance
```

**Variations:**
- Co-first authors (equal contribution): marked with *
- Co-corresponding authors
- Alphabetical order (in some fields)

**Best Practice:**
- ‚úÖ Discuss authorship early
- ‚úÖ Document contributions
- ‚úÖ Use CRediT taxonomy (Contributor Roles)
- ‚úÖ Resolve disputes before submission

---

### üôè Acknowledgments Section

**Who goes here:**
- ‚úÖ Provided funding
- ‚úÖ Provided equipment/facilities
- ‚úÖ Technical assistance
- ‚úÖ General advice
- ‚úÖ Manuscript editing (but not writing)
- ‚úÖ Helped with fieldwork

**Example:**
```
Acknowledgments

We thank the Odisha State Higher Education Council (OSHEC) for funding 
(Grant No. XXXX). We are grateful to Dr. X for assistance with statistical 
analysis, Mr. Y for help with field collections, and the Department of 
Zoology, Kuchinda College for providing laboratory facilities. We thank 
anonymous reviewers for constructive comments that improved this manuscript.
```

## 7.5 Conflicts of Interest

### ‚öñÔ∏è Disclosure Requirements

**What is a Conflict of Interest (COI)?**
- Situation where personal interests could bias judgment
- Financial, personal, or professional

**Types:**

#### **Financial COI**
- Funding from company with stake in results
- Stock ownership in relevant company
- Consulting fees
- Patent/royalty interests

#### **Non-Financial COI**
- Personal relationships
- Professional competition
- Strongly held beliefs

---

### üìù How to Handle COI

**1. Disclose**
- Always declare potential COI
- Journals have disclosure forms
- Better to over-disclose than under-disclose

**2. Manage**
- Recuse from decisions
- Have independent data analysis
- Extra scrutiny from collaborators

**3. Avoid When Possible**
- Don't accept funding if compromises integrity
- Don't review competitor's papers

**Example Disclosure Statement:**

## 8.1 Animal Research Ethics

### üêÅ The 3Rs Principle

**Framework for Humane Animal Research:**

#### **1. Replace**
**Use alternatives when possible**

**Alternatives:**
- ‚úÖ Computer simulations
- ‚úÖ In vitro cell culture
- ‚úÖ Mathematical models
- ‚úÖ Studies using non-sentient organisms
- ‚úÖ Observational studies (no manipulation)

**When animals ARE necessary:**
- Complex whole-organism responses needed
- Behavior studies
- No suitable alternative exists

---

#### **2. Reduce**
**Use minimum number necessary**

**How to Reduce:**
- ‚úÖ Proper statistical power analysis
- ‚úÖ Efficient experimental design
- ‚úÖ Share tissues/data with other researchers
- ‚úÖ Use animals that die naturally when possible
- ‚úÖ Better data analysis (extract more info from each animal)

**DON'T:**
- ‚ùå Use too few (underpowered study)
- ‚ùå Use too many (waste)
- ‚ùå Duplicate existing work unnecessarily

---

#### **3. Refine**
**Minimize pain and distress**

**Refinements:**
- ‚úÖ Use anesthesia/analgesia
- ‚úÖ Proper housing (space, enrichment, social groups)
- ‚úÖ Gentle handling
- ‚úÖ Humane endpoints (euthanize before severe suffering)
- ‚úÖ Staff training
- ‚úÖ Health monitoring

---

### üìã CPCSEA Guidelines (India)

**Committee for the Purpose of Control and Supervision of Experiments on Animals**

**Key Requirements:**

1. **Institutional Animal Ethics Committee (IAEC)**
   - Required for all animal research
   - Reviews and approves protocols
   - Monitors compliance

2. **Registration**
   - Institution must be registered with CPCSEA
   - Regular inspections

3. **Protocol Requirements**
   - Scientific justification
   - 3Rs implementation
   - Humane endpoints
   - Veterinary care plan

### üêü Special Considerations for Fish

**Fish ARE animals and deserve ethical consideration!**

**Common Misconception:**
- ‚ùå "Fish don't feel pain"
- ‚úÖ Evidence shows fish experience pain and stress

**Ethical Fish Research:**
- ‚úÖ Appropriate water quality (temperature, pH, DO)
- ‚úÖ Adequate space
- ‚úÖ Minimize handling stress
- ‚úÖ Anesthesia for invasive procedures (MS-222, clove oil)

**Euthanasia Methods:**

**Acceptable:**
- ‚úÖ Anesthetic overdose (MS-222, benzocaine)
- ‚úÖ Rapid chilling (ice water for small fish)
- ‚úÖ Decapitation (after anesthesia)

**NOT Acceptable:**
- ‚ùå Suffocation in air
- ‚ùå Freezing
- ‚ùå Carbon dioxide exposure

## 8.4 Institutional Review Boards (IRBs)

### üìã Ethics Review Process

**What is an IRB/IEC?**
- Institutional Review Board (USA) / Institutional Ethics Committee (India)
- Reviews research proposals
- Ensures ethical compliance
- Protects participants

**When Required:**
- Animal research (IAEC)
- Human subjects research (IEC)
- Use of biological materials

**Review Levels:**

1. **Exempt**
   - Minimal risk
   - No vulnerable populations
   - Example: Anonymous surveys about wildlife preferences

2. **Expedited**
   - Minimal risk
   - Reviewed by chair or designated member
   - Example: Non-invasive fish observation

3. **Full Board Review**
   - More than minimal risk
   - Vulnerable populations
   - Novel procedures
   - Example: Experimental treatments on animals

## 8.5 Making Ethical Decisions

### ü§î Ethical Decision-Making Framework

**When Facing Ethical Dilemma:**

**1. Identify the Issue**
- What is the ethical question?
- Who is affected?
- What are the stakes?

**2. Gather Information**
- What are the facts?
- What are the relevant guidelines/laws?
- What have others done in similar situations?

**3. Consider Alternatives**
- What are your options?
- What are consequences of each?

**4. Apply Ethical Principles**
- Respect for persons?
- Beneficence (do good)?
- Non-maleficence (do no harm)?
- Justice (fairness)?

**5. Consult**
- Talk to supervisor
- Ethics committee
- Trusted colleagues

**6. Decide and Document**
- Make decision
- Document reasoning
- Implement
- Reflect on outcome

### üí° Common Ethical Scenarios

**Scenario 1:** Supervisor asks you to add their name to paper they didn't contribute to
- **Ethical Issue:** Gift authorship
- **What to do:** Politely explain authorship criteria, suggest acknowledgment instead

**Scenario 2:** Data doesn't support hypothesis, supervisor suggests excluding outliers
- **Ethical Issue:** Potential falsification
- **What to do:** Ask for clear justification, document decision, report truthfully

**Scenario 3:** Accidentally discover threatened species population during fieldwork
- **Ethical Issue:** Obligation to report vs. research goals
- **What to do:** Report to authorities, protect location, contribute to conservation

**Scenario 4:** Local community asks for compensation beyond agreed amount
- **Ethical Issue:** Fair compensation vs. budget constraints
- **What to do:** Negotiate openly, consider indirect benefits (training, results sharing)
